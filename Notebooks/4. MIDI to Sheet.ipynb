```python
from music21 import converter, corpus

def midi_to_sheet(input, output):
    # Specify the paths for your input MIDI file and output MuseScore file
    midi_file_path = input
    mscz_file_path = output

    # Convert the MIDI file to a Music21 Score object
    score = converter.parse(midi_file_path)

    # Save the Music21 Score as a MuseScore file
    score.write('musicxml', mscz_file_path)

    print(f"Conversion complete. The MuseScore file is saved at: {mscz_file_path}")
```

```python
from music21 import note
from music21 import instrument

def display_sheet(input, instruments):
    littleMelody = converter.parse(input)

    # Iterate through the parts (tracks) and assign instruments
    for i, part in enumerate(littleMelody.parts[:-1]):
        part.insert(0, instruments[i])  # Assign the instrument to the beginning of the track
    instr = []
    # Iterate through the notes in first part
    for note in littleMelody.parts[0].recurse().notes:
        note.addLyric("a")  # Add lyrics to each note
        
        
    print(instr)
        
    
    littleMelody.show()
```
