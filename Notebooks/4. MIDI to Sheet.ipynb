```python
from music21 import converter

def midi_to_sheet(input, output):
    midi_file_path = input
    mscz_file_path = output
    score = converter.parse(midi_file_path)
    score.write('musicxml', mscz_file_path)

    print(f"Conversion complete. The MuseScore file is saved at: {mscz_file_path}")
```

```python
from music21 import instrument

def display_sheet(input, instruments):
    littleMelody = converter.parse(input)
    for i, part in enumerate(littleMelody.parts[:-1]):
        part.insert(0, instruments[i])
    littleMelody.show()
```

```python
midi_to_sheet("sample/20191213_Bella_Ciao.mid", "sample/20191213_Bella_Ciao.mscz")
```

```python

instruments = [
    instrument.Vocalist(),
    instrument.Soprano(),  # Soprano
    instrument.MezzoSoprano(),
    instrument.Alto(),  # Alto
    instrument.Tenor(),  # Tenor
    instrument.Tenor(),  # Bass
]

display_sheet("sample/20191213_Bella_Ciao.mid", instruments)
```
